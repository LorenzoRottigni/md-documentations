# 07 pandas data visualziation

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df2 = pd.read_csv('df2')
```

```python
df1.head()
df2.head()
```

```python
# Histogram
df1['A'].hist(bins=30)
```

```python
# df columns expose plot method
df1['A'].plot(
    # what kind of plot
    kind='hist',
    bins=30
)
# shorthand
df1['A'].plot.hist(bins=30)
```

```python
df2.plot.area(
    # transparency
    alpha=0.4
)
```

```python
# considers categorical indexes, if number they will be treated as categories
df2.plot.bar()
```

```python
# onlw shows the offset between bars
df2.plot.bar(stacked=True)
```

```python
df1.plot.line(x=df1.index, y='B')
```

```python
df1.plot.scatter(
    # feature A on x axis
    x='A',
    # feature B on y axis
    y='B',
    # color based on feature C, adds 3rd dimension
    c='C',
    # adjust size of dots based on feature C
    s=df1['C'] * 100,
    cmap='coolwarm'
)
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a', y='b', gridsize=25, cmap='coolwarm')
```

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
