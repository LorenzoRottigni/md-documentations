```python
import seaborn as sns
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.pairplot(tips, hue="sex", palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

## CATEGORICAL PLOTS

```python
import numpy as np
sns.barplot(
    # feature sex on X
    x='sex',
    # feature total_bill on Y
    y='total_bill',
    data=tips,
    # estimator is the function to aggregate the data, by default is mean
    # np.std is the standard deviation
    estimator=np.std
)
```

```python
# similar to barplot, but the numerical parameter is the occurrences count
sns.countplot(x='sex',data=tips)
```

```python
# distribution of total_bill belonging to each day
sns.boxplot(
    # categorical feature
    x='day',
    # numerical feature
    y='total_bill',
    data=tips,
    # specify another categorical feature to split the data
    hue='smoker'
)
```

```python
# allows to understand the relationship between two categorical features and a numerical feature
sns.violinplot(
    # categorical feature
    x='day',
    # numerical feature
    y='total_bill',
    data=tips,
    hue='sex',
    # split the violin plot by the hue feature
    # instead of having a violin plot for each category of the hue feature
    split=True
)
```

```python
sns.stripplot(
    # categorical feature
    x='day',
    # numerical feature
    y='total_bill',
    data=tips,
    # adds a random noise to the data to avoid overlapping
    jitter=True,
    hue='sex'
)
```

```python
# a combination of stripplot and violinplot
sns.swarmplot(
    x='day',
    y='total_bill',
    data=tips,
)
```

```python
sns.factorplot(
    x='day',
    y='total_bill',
    data=tips,
    # specify the kind of plot
    kind='bar'
)
```
